```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
distance = [0, 2, 4, 6, 8, 10]  # Positions along the beam (meters)
shear_force = [10, 5, 0, -5, -10, 0]  # Shear force at each position (kN)
bending_moment = [0, 10, 15, 10, 5, 0]  # Bending moment at each position (kN.m)

# Plot Shear Force Diagram (SFD)
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(distance, shear_force, marker='o', linestyle='-', color='b', label="Shear Force")
plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel("Distance (m)")
plt.ylabel("Shear Force (kN)")
plt.title("Shear Force Diagram (SFD)")
plt.grid(True)
plt.legend()

# Plot Bending Moment Diagram (BMD)
plt.subplot(2, 1, 2)
plt.plot(distance, bending_moment, marker='o', linestyle='-', color='r', label="Bending Moment")
plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel("Distance (m)")
plt.ylabel("Bending Moment (kN.m)")
plt.title("Bending Moment Diagram (BMD)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
